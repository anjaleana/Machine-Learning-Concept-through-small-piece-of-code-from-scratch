Matplotlib is visiualization technique. It lies under the pyplot submodule

#### 1st method

```python
import matplotlib.pyplot as plt
import numpy as np
xtrain = np.array([0,10])
ytain = np.array([0,500])
plt.plot(xtrain,ytain)
plt.show()
```

Observation : The x-axis represents the xtrain and y-axis represents the ytrain. the plot function plots the line from one point to another point on both x and y- axis

#### 2nd method

```python
plt.plot(xtrain,ytain,'o')
plt.show()
```

#### 3rd method

```python
xtrain = np.array([1,3,6,7])
ytrain = np.array([3,8,1,10])
plt.plot(xtrain,ytrain)
plt.show()
```

#### 4th method

```python
ytrain = np.array([3,4,2,10,17,19,20,25])
plt.plot(ytrain)
plt.show()
```

Observation: If we don't defined the x-axis values based on the length of y-axis values, it takes on x-axis and draw the line

#### If you want to emphasize the specified point we use marker keyword(5th method)

```python
plt.plot(ytrain,marker='o')
plt.show()
```

#### 6th method

```python
plt.plot(ytrain,marker='*')
plt.show()
```

#### 7th method

```python
plt.plot(ytrain,'o:r')
plt.show()
```

```python
plt.plot(ytrain,'o-g')
plt.show()
```

#### 8th method

#### Using marker ,we can also add marker size with short version ms

```python
plt.plot(ytrain,marker='o',ms=20)
plt.show()
```

#### Using marker ,marker_size, we can also use marker edge color . We user short format for markeredgecolor mec

```python
plt.plot(ytrain,marker='o',ms=20,mec='r')
plt.show()
```

#### We can use markerfacecolor or mfc to set the color inside the edge of markers

```python
plt.plot(ytrain,marker='o',ms=20,mfc='r')
plt.show()
```
